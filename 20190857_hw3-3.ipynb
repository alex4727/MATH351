```python
from numpy import *
from pylab import *
```

```python
def g(x):
    tmp = 5 / (x ** 0.5)
    return tmp
```

```python
p0 = 2.5
for i in range(100):
    p1 = g(p0)
    p0 = p1
print(g(p1)-p0)
print(p0)
```

```python
a,b = 2.5, 3.5
x = linspace(a,b, 1000); y0=x ; y=g(x)
plot(x, y0, '-b', x, y, '-r')
```

```python
%pylab
p0 = 2.5; plot(p0,p0,'go')
for i in range(10):
    p1 = g(p0)
    plot([p0, p0], [p0, p1], '-k')
    pause(0.5)
    plot([p0, p1], [p1, p1], '-k')
    pause(0.5)
    p0 = p1
```
