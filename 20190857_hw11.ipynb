```python
import numpy as np
from math import comb
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
## hw 11

n = 5
x = np.linspace(0, 1, 100000)
def get_y(n, x, k):
    return comb(n,k) * (x**k) * (1-x)**(n-k)
y_0 = get_y(n, x, 0)
y_1 = get_y(n, x, 1)
y_2 = get_y(n, x, 2)
y_3 = get_y(n, x, 3)
y_4 = get_y(n, x, 4)
y_5 = get_y(n, x, 5)

plt.plot(x, y_0, 'r')
plt.plot(x, y_1, 'g')
plt.plot(x, y_2, 'b')
plt.plot(x, y_3, 'cyan')
plt.plot(x, y_4, 'brown')
plt.plot(x, y_5, 'purple')
```

```python
def f(x):
    return np.exp(x)

def get_bernstein(n, x):
    tmp = 0
    for k in range(0, n+1):
        tmp += get_y(n, x, k) * f(k/n)
    return tmp

x = 0.5
errors = []
ns = []
for i in range(1, 10):
    ns.append(i)
    errors.append(np.log(abs(f(x)-get_bernstein(2**i, x))))
plt.plot(ns, errors, "-o")
```

```python
exp_ns = [2 ** i for i in ns]
plt.plot(exp_ns, errors, "-o")
```
