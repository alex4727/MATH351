```python
pa = [1]
pb = [1]
pc = [1]
pd = [1]

for i in range(10):
    if i < 3:
        new_pa = pa[i] * (1+(7-pa[i]**5)/pa[i]**2)**3
        pa.append(new_pa)
    if i < 5:
        new_pb = pb[i] - (pb[i]**5-7)/(pb[i]**2)
        pb.append(new_pb)
        
    new_pc = pc[i] - (pc[i]**5-7)/(5*pc[i]**4)
    new_pd = pd[i] - (pd[i]**5-7)/12
    pc.append(new_pc)
    pd.append(new_pd)
    
print(f"Actual value: {7**(1/5)}")
print(f"Pa sequence: {pa}")
print(f"Pb sequence: {pb}")
print(f"Pc sequence: {pc}")
print(f"Pd sequence: {pd}")
```
