```python
from numpy import *
from pylab import *
%pylab
```

```python
def f(x):
    tmp = x**2 - 10*cos(x)
    return tmp    
def fp(x):
    tmp = 2*x + 10*sin(x)
    return tmp
```

```python
index = 0
p = 1
err = 1
while err >=10**-8:
    index = index + 1
    pold = p
    p = p - f(p)/fp(p)
    print(index, p)
    # plot(x, fx,'-r', p, f(p), 'bo', markersize=10)
    err = abs(pold - p)
    # pause(1)
print(p)
```
